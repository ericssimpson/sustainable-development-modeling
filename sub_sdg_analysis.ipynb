```python
import igraph as ig
sdg_bipartite_ig = ig.Graph.Read(
    f='sdg_and_sub_bipartite.gml',
    format='gml'
    )
import networkx as nx
sdg_bipartite_nx = nx.read_gml(
    path='sdg_and_sub_bipartite.gml'
    )
import netgraph as ng
```

```python
top = nx.bipartite.sets(sdg_bipartite_nx)[0]
pos = nx.bipartite_layout(sdg_bipartite_nx, top)
nx.draw_networkx(sdg_bipartite_nx, pos=pos)
```

```python
import networkx.algorithms.community as nx
sdg_clusters_nx = nx.greedy_modularity_communities(sdg_bipartite_nx)
```

```python
sdg_clusters_ig = sdg_bipartite_ig.community_fastgreedy()
ig.plot(sdg_clusters_ig)
```
