```python
import igraph as ig
sdg_bipartite_ig = ig.Graph.Read(
    f='sdg_and_sub_bipartite.gml',
    format='gml'
    )
sdg_complete_ig = ig.Graph.Read(
    f='sdg_complete.gml',
    format='gml'
    )
import networkx as nx
sdg_bipartite_nx = nx.read_gml(
    path='sdg_and_sub_bipartite.gml'
    )
sdg_complete_nx = nx.read_gml(
    path='sdg_complete.gml'
    )
```

```python
from community import community_louvain
complete_sdg_clusters_nx = community_louvain.best_partition(sdg_complete_nx)
print(complete_sdg_clusters_nx)
bipartite_sdg_clusters_nx = community_louvain.best_partition(sdg_bipartite_nx)
```

```python
import netgraph as ng
community_to_color = {
    0 : 'tab:blue',
    1 : 'tab:orange',
    2 : 'tab:green',
    3 : 'tab:red',
}
```

```python
node_color_1 = {node: community_to_color[community_id] for node, community_id in complete_sdg_clusters_nx.items()}
ng.Graph(sdg_complete_nx,
    node_color=node_color_1, node_edge_width=0, edge_alpha=0.5,
    node_layout='community', node_layout_kwargs=dict(node_to_community=complete_sdg_clusters_nx)
)
```

```python
node_color_2 = {node: community_to_color[community_id] for node, community_id in bipartite_sdg_clusters_nx.items()}
ng.Graph(sdg_bipartite_nx,
    node_color=node_color_2, node_edge_width=0, edge_alpha=0.5,
    node_layout='community', node_layout_kwargs=dict(node_to_community=bipartite_sdg_clusters_nx)
)
```
