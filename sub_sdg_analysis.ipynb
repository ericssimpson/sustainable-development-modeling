```python
from igraph import *
import networkx as nx
```

```python
sdg_bipartite = nx.read_gml(
    path="sdg_and_sub_bipartite.gml"
    )
```

```python
nx.write_graphml(sdg_bipartite, 'sdg_and_sub_bipartite.graphml')
```

```python
top = nx.bipartite.sets(sdg_bipartite)[0]
pos = nx.bipartite_layout(sdg_bipartite, top)
```

```python
nx.draw(sdg_bipartite, pos=pos)
```
